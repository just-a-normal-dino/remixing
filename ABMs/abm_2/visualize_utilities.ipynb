```python
def plot_repo(Repo, x_init=None, x_final=None):
    x = []
    y = []
    for design in Repo:
        x.append(design[1])
        y.append(design[0])
    plt.scatter(y,x, marker='.')

    if not x_init==None:
        plt.scatter(x_init[0],x_init[1], marker='x', label= 'initial design')
    if not x_final==None:
        plt.scatter(x_final[0],x_final[1], marker='x', label= 'final design')
```

```python
def Repo_histogram(model, x_final=None, title=None):
    avg_utils = [model.avg_utility(x) for x in model.Repo]
    plt.hist(x=avg_utils, bins=20, density=True)
    if x_final != None:
        plt.scatter(model.avg_utility(x_final), 0, color='red', label='final design')
    if title != None:
        plt.title(title)
    else:
        plt.title('Histogram of avg utility in Repo')
    plt.xlabel('avg utility')
    plt.ylabel('num designs')
    plt.legend()
    plt.show()

def display_heatmap(image, title=None):
    """
    Display the input image as a heatmap.

    Parameters:
        image (numpy.ndarray): Input image.
        title (str): Title for the plot.
    """
    plt.imshow(np.transpose(image), cmap='hot', vmin=0, vmax=1)
    plt.colorbar()
    if not title==None:
        plt.title(title)
    plt.show()

```

### Visualisation of the utility functions

```python
m100 = Simulation(numAgents=100, diversity=0.5, utility_parameter=30)
display_heatmap(m100.Uavg, title='average utility for 100 agents')
```

```python
m500 = Simulation(numAgents=500, diversity=0.5, utility_parameter=30)
display_heatmap(m500.Uavg, title='average utility for 500 agents')
```
